## Exercises using matplotlib

1. Use matplotlib to plot the following equation:

    $y = x^2 - x + 2$

    Add an anotation for the point 0, 0, the origin.

First, import the needed libraries.

```python
# adds interactive graphs, be sure to put before import
# matplotlib notebook

import matplotlib.pyplot as plt
import math

%matplotlib inline
```

Then, use `plt` functions to create the plot.
- The `.text` function creates an arrow, with `xy` being the point of the arrow and `xytext` being the end of it.

```python
x = range(-5, 6)
y = [n ** 2 - n + 2 for n in x]
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 2.7, '(0,0)')
plt.show()
```

2. Create and label 4 separate plots for the following equations (choose a range for x that makes sense):

- $y = √x$
- $y = x^3$
- $y = tan(x)$
- $y = 2^x$

    The math module is helpful here.

$y = √x$

```python
x = list(range(0, 8))
y = [math.sqrt(n) for n in x]
plt.title('$√x$')
plt.plot(x, y)
plt.show()
```

$y = x^3$

```python
x = list(range(-7, 8))
y = [n ** 3 for n in x]
plt.title('$x^3$')
plt.plot(x, y)
plt.show()
```

$y = tan(x)$

```python
x = list(range(0, 8))
y = [math.tan(n) for n in x]
plt.title('$tan(x)$')
plt.plot(x, y)
plt.show()
```

$y = 2^x$

```python
x = list(range(-7, 8))
y = [2 ** n for n in x]
plt.title('$2^x$')
plt.plot(x, y)
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (12, 8)) # (width, height)
plt.suptitle('All graphs')

x1 = list(range(0, 8))
y1 = [math.sqrt(n) for n in x1]
plt.subplot(221)
plt.title('$√x$')
plt.plot(x1, y1)

x2 = list(range(-7, 8))
y2 = [n ** 3 for n in x2]
plt.subplot(222)
plt.title('$x^3$')
plt.plot(x2, y2)

x3 = list(range(0, 8))
y3 = [math.tan(n) for n in x3]
plt.subplot(223)
plt.title('$tan(x)$')
plt.plot(x3, y3)

x4 = list(range(-7, 8))
y4 = [2 ** n for n in x4]
plt.subplot(224)
plt.title('$2^x$')
plt.plot(x4, y4)

plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
plt.figure(figsize = (12, 8))

plt.plot(x1, y1, label = '$√x$')

plt.plot(x2, y2, label = '$n^3$')

plt.plot(x3, y3, label = '$tan(x)$')

plt.plot(x4, y4, label = '$2^n$')

plt.title('All graphs in one plot')
plt.legend(loc = 'best')

plt.show()
```
